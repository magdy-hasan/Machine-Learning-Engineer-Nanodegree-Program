```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
da = pd.read_csv('data1.txt',sep="\t")
```

```python
da = da[(da['gender'] == 'm') | (da['gender'] == 'f')]
```

```python
da['gender'].value_counts()
```

```python
def create_path(r):
    return r['user_id'] + '/landmark_aligned_face.' + str(r['face_id']) + '.' + r['original_image']
```

```python
da['path'] = da.apply(create_path, axis=1)
```

```python
da
```

```python
def v_img(i):
    print(da['gender'][i])
    plt.imshow(plt.imread('aligned/' + da['path'][i] ))
```

```python
v_img(15)
```

```python
da = da[['gender','path']]
```

```python
da
```

```python
da = da.sample(frac=1).reset_index(drop=True)
```

```python
dtm = da[da.gender == 'm'].sample(n=5000)
dtf = da[da.gender == 'f'].sample(n=5000)
```

```python
dtr = dtm[0:3000].append(dtf[0:3000])
dval = dtm[3000:4000].append(dtf[3000:4000])
dte = dtm[4000:].append(dtf[4000:])
```

```python
dtr = dtr.sample(frac=1).reset_index(drop=True)
dval = dval.sample(frac=1).reset_index(drop=True)
dte = dte.sample(frac=1).reset_index(drop=True)
```

```python
pickle.dump([dtr,dval,dte],open('gender_data.pkl','wb'))
```

```python
dtr,dval,dte = pickle.load(open('gender_data.pkl','rb'))
```

```python
print(len(dtr),len(dval),len(dte))
```

```python
dte['gender'].value_counts().plot.bar()
```

```python
from keras.preprocessing import image
from keras.applications.inception_resnet_v2 import preprocess_input
from keras.applications import InceptionResNetV2
from keras.models import Model
import pickle
```

```python
model = InceptionResNetV2()
```

```python
model.layers.pop()
model = Model(model.inputs,model.layers[-1].output)
```

```python
len(dtr)
```

```python
X,Y = [],[]
cnt = 0
for i,r in dtr.iterrows():
    path = 'aligned/' + r['path']
    img = image.load_img(path, target_size=(299,299))
    img = image.img_to_array(img)
    img = img.reshape(1,img.shape[0],img.shape[1],img.shape[2])
    img = preprocess_input(img)
    X.append(model.predict(img)[0])
    if i % 500 == 0:
        print(i)
```

```python
x = np.array(X)
```

```python
x.shape
```

```python
y = dtr['gender'] == 'm'
y = y.astype(int)
```

```python
y.shape
```

```python
pickle.dump([x,y],open('tri2.pkl','wb'))
```

```python
del x,y
```
