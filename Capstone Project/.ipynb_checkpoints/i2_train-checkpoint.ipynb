```python
from keras.models import Model
from keras import layers
import pickle
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.applications import VGG16
from keras.callbacks import ModelCheckpoint
%matplotlib inline
```

```python
# load the data we created in pre_data_inc2.ipynb
x,y = pickle.load(open('tri2.pkl','rb'))
xv,yv = pickle.load(open('vali2.pkl','rb'))
```

```python
# check the size of the data
print(x.shape,y.shape)
print(xv.shape,yv.shape)
```

### create a simple network with one dense layer and dropout layer 

```python
# input features
in1 = layers.Input(shape=(1536,))

# a fully connected layer with dropout
X = layers.Dense(1024,activation='relu')(in1)
X = layers.Dropout(0.8)(X)

# final layer
X = layers.Dense(1,activation='sigmoid')(X)

model = Model(inputs=in1,outputs=X)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
# save best model of validtion accuracy in i2_saved_models folder
filepath = 'i2_saved_models\model-ep{epoch:03d}-acc{acc:.3f}-val_acc{val_acc:.3f}.h5'
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=0, save_best_only=True, mode='max')
```

```python
# train the model for 300 epoochs with batch size 512
hist = model.fit(x,y,epochs=300,batch_size=512,validation_data=[xv,yv],verbose=2,callbacks=[checkpoint])
```

```python
# summarize history for accuracy
plt.plot(hist.history['acc'])
plt.plot(hist.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### testing the model on new images

```python
from keras.applications import InceptionResNetV2
from keras.preprocessing import image
from keras.applications.inception_resnet_v2 import preprocess_input
from keras.models import Model
from keras.models import load_model
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
# model for extracting features
bmodel = InceptionResNetV2()
```

```python
bmodel.layers.pop()
bmodel = Model(bmodel.inputs,bmodel.layers[-1].output)
```

```python
# model for predicting given extracted features
model = load_model('i2_saved_models/model-ep236-acc0.931-val_acc0.889.h5')
```

```python
def pre(path):
    # given image path, predict gender
    
    # show the image
    plt.imshow(plt.imread(path))
    # load image with size (299,299,3)
    img = image.load_img(path, target_size=(299,299))
    img = image.img_to_array(img)
    # reshape to (#sample,299,299,3)
    img = img.reshape(1,img.shape[0],img.shape[1],img.shape[2])
    img = preprocess_input(img)
    # compare model output to some threshold ( here i choose 0.45 )
    if model.predict(bmodel.predict(img))[0][0] > 0.45:
        print('male')
    else:
        print('female')
    return
```
