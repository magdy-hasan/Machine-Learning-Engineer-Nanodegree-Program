### import needed libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

### load data from data.txt file

```python
da = pd.read_csv('data.txt',sep="\t")
```

### view data distribution of gender column

```python
plt.ylabel('count of images in the dataset')
plt.xlabel('gender label')
da['gender'].value_counts().plot.bar()
```

```python
#select data with gender label 'm' or 'f' only
da = da[(da['gender'] == 'm') | (da['gender'] == 'f')]
```

```python
#view columns of the dataframe
da.columns
```

### create  new column 'path' that contain path to images in 'aligned' folder 

```python
def create_path(r):
    return r['user_id'] + '/landmark_aligned_face.' + str(r['face_id']) + '.' + r['original_image']
```

```python
da['path'] = da.apply(create_path, axis=1)
```

```python
#select only gender and path columns as we only intersted in this two
da = da[['gender','path']]
```

```python
#reset indices of the datafrane and randomize it
da = da.sample(frac=1).reset_index(drop=True)
```

### explore some images

```python
def view_img(i):
    print(da['gender'][i])
    plt.imshow(plt.imread('aligned/' + da['path'][i] ))
```

### sample 1000 images of the 17492 such that 5000 of them with 'f' label and 5000 with 'm' label

```python
dtm = da[da.gender == 'm'].sample(n=5000)
dtf = da[da.gender == 'f'].sample(n=5000)
```

### from 10000 we will use 6000 as training data and 2000 as validation data and 2000 as test data/]

```python
dtr = dtm[0:3000].append(dtf[0:3000])
dval = dtm[3000:4000].append(dtf[3000:4000])
dte = dtm[4000:].append(dtf[4000:])
```

```python
# reset incides of all tables 
dtr = dtr.sample(frac=1).reset_index(drop=True)
dval = dval.sample(frac=1).reset_index(drop=True)
dte = dte.sample(frac=1).reset_index(drop=True)
```

```python
# save them to use them in different experiments
pickle.dump([dtr,dval,dte],open('gender_data.pkl','wb'))
```
